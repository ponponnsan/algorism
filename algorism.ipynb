```python
# バブルソート
list = [9,8,7,6,5,4,3,2,1]
n = len(list)
i=0
while i < n:
#     比較する順番は後ろから
    j = n-1
    
    while j > i:
#         交換する値は隣同士
        if list[j] < list[j-1]:
            temp = list[j]
            list[j] = list[j-1]
            list[j-1] =temp
            
            print(list)
        j= j-1
    i = i+1
    print('何回目の配列か：',i)
```

```python
# 選択ソート
list = [9,2,7,4,6,5,1,3,8]
n = len(list)
i=0
while i < n:
    min_idx = i
#     繰り返し処理を制御する値
    j = i
#     未整列の配列部分から値を選ぶ
    while j < n:
        if list[min_idx] > list[j]:
            temp = list[min_idx]
            list[min_idx] = list[j]
            list[j] = temp
            
            print(list)
        j = j+1
    i = i+1
    print('何回目の配列か：',i)
```

```python
# 挿入ソート
list = [9,8,7,6,5,4,3,2,1]
n = len(list)
i=0
# iは整列済み配列の一番後ろの数字
while i < n:
    j = 0
#     選んだ値を整列済みのどこに挿入するか探す
    while j < i:
        if list[i] < list[j]:
            temp = list[i]
            list[i] = list[j]
            list[j] = temp
            print(list)
        j = j+1
    i = i+1
    print('何回目の配列か：',i)
    
```
